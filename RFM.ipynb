Recency

To calculate recency, we need to choose a date point from which we evaluate how many days ago was the customer's last purchase.

```python
dfnew['order date (DateOrders)'].max()
```

```python
import datetime as dt
refdate = dt.date(2018,1,31)
print(refdate)
```

```python
dfnew['date'] = dfnew['order date (DateOrders)'].dt.date
```

```python
dfnew.shape
```

```python
recency_df = dfnew.groupby(by='Customer Id', as_index=False)['date'].max()
recency_df.columns = ['Customer Id','LastPurshaceDate']
recency_df.head()
```

```python
#calculate recency
recency_df['Recency'] = recency_df['LastPurshaceDate'].apply(lambda x: (refdate - x).days)
```

```python
recency_df.head()
```

```python
recency_df.drop('LastPurshaceDate',axis=1,inplace=True)
```

Frequency

Frequency helps us to know how many times a customer purchased from us. To do that we need to check how many invoices are registered by the same customer.

```python
# drop duplicates
dfnew_copy = dfnew
dfnew_copy.drop_duplicates(subset=['Order Id', 'Customer Id'], keep="first", inplace=True)
#calculate frequency of purchases
frequency_df = dfnew_copy.groupby(by=['Customer Id'], as_index=False)['Order Id'].count()
frequency_df.columns = ['Customer Id','Frequency']
frequency_df.head()
```

```python
frequency_df['Frequency'].value_counts()
```

Monetary

Monetary attribute answers the question: How much money did the customer spent over time?
To do that, first, we will create a new column total cost to have the total price per invoice.

```python
#create column total cost
dfnew['TotalCost'] = dfnew['Order Item Quantity'] * dfnew['Product Price']
```

```python
monetary_df = dfnew.groupby(by='Customer Id',as_index=False).agg({'TotalCost': 'sum'})
monetary_df.columns = ['Customer Id','Monetary']
monetary_df.head()
```

```python
temp_df = recency_df.merge(frequency_df,on='Customer Id')
temp_df.head()
```
